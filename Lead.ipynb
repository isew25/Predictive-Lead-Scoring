```python
def Describe(data):
    total = data.isnull().sum().sort_index()
    percent = (data.isnull().sum()/data.isnull().count()*100).sort_index()
    unique = data.nunique().sort_index()

    value = []
    for i in data.columns.sort_values():
      value.append(data[i].unique())
      
    D = pd.DataFrame({'Unique value' : value},index=data.columns.sort_values())
    
    return pd.concat([data.dtypes.sort_index(), total, percent, unique, D['Unique value']], axis=1, keys=['Dtype','Missing value', 'Missing value (%)','N-Unique','Unique value']).sort_values(by='Missing value',ascending = False)
```

### Modelling

```python
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction import DictVectorizer


X = data.loc[:,data.columns != 'Converted']
y = data['Converted']

dv = DictVectorizer(sparse=False)

X_dict = X.to_dict(orient='rows')
X = dv.fit_transform(X_dict)
X = pd.DataFrame(X,columns=dv.get_feature_names())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import cross_validate
from sklearn.model_selection import StratifiedKFold


rs = 12
model = [LogisticRegression(solver='liblinear', random_state = rs), 
         RandomForestClassifier(random_state = rs),
         DecisionTreeClassifier(random_state = rs),
         GradientBoostingClassifier(random_state = rs),
         lgb.LGBMClassifier(learning_rate=0.09,max_depth=-12,random_state = rs),
         xgb.XGBClassifier(random_state = rs)
         ]
model_name = ['Logstic Regression', 'Random Forest', 'Dessicion Tree', 'Gradient Boosting', 'LGBM', 'XGBoost']


def clf_model(X,y,model,model_name):

    skf = StratifiedKFold(n_splits=10)

    M_cv = {}
    for i in range(len(model)):
      M_cv[model_name[i]] = cross_validate(model[i], X, y, cv=skf,scoring=['roc_auc','accuracy','f1','recall','precision'])

    #index = ['Fold1','Fold2','Fold3', 'Fold4','Fold5','Fold6', 'Fold7','Fold8','Fold9', 'Fold10']
    #M_cv = pd.DataFrame(M_cv,index=index).T
    #M_cv['mean'] = M_cv.mean(axis=1)
    #M_cv['std'] = M_cv.std(axis=1)
    D = {}

    for i in list(M_cv.keys()):
      D[i] = list()
      for j in list(M_cv[i].keys())[2:]:
        D[i].append(M_cv[i][j].mean())
    return pd.DataFrame(D,index=list(M_cv[i].keys())[2:]).T.sort_values(by='test_roc_auc', ascending=False).style.highlight_max(color = 'red', axis = 0)
```

```python
M = clf_model(X,y,model,model_name)
M
```

```python
from bayes_opt import BayesianOptimization
from skopt  import BayesSearchCV 
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
%%time

def bayes_parameter_opt_lgb(X, y, init_round, opt_round, n_folds, random_seed,n_estimators, output_process=False):
    # prepare data
    train_data = lgb.Dataset(data=X, label=y, free_raw_data=False)
    # parameters
    def lgb_eval(learning_rate,num_leaves, feature_fraction, bagging_fraction, max_depth, max_bin, min_data_in_leaf,min_sum_hessian_in_leaf,subsample):
        params = {'application':'binary', 'metric':'auc'}
        params['learning_rate'] = max(min(learning_rate, 1), 0)
        params["num_leaves"] = int(round(num_leaves))
        params['feature_fraction'] = max(min(feature_fraction, 1), 0)
        params['bagging_fraction'] = max(min(bagging_fraction, 1), 0)
        params['max_depth'] = int(round(max_depth))
        params['max_bin'] = int(round(max_depth))
        params['min_data_in_leaf'] = int(round(min_data_in_leaf))
        params['min_sum_hessian_in_leaf'] = min_sum_hessian_in_leaf
        params['subsample'] = max(min(subsample, 1), 0)
        
        cv_result = lgb.cv(params, train_data, nfold=n_folds, seed=random_seed, stratified=True, verbose_eval =200, metrics=['auc'])
        return max(cv_result['auc-mean'])
     
    lgbBO = BayesianOptimization(lgb_eval, {'learning_rate': (0.001, 1.0),
                                            'num_leaves': (4, 100),
                                            'feature_fraction': (0.1, 0.9),
                                            'bagging_fraction': (0.5, 1),
                                            'max_depth': (5, 70),
                                            'max_bin':(10,90),
                                            'min_data_in_leaf': (10, 80),
                                            'min_sum_hessian_in_leaf':(0,100),
                                            'subsample': (0.01, 1.0)}, random_state=200)


    
    lgbBO.maximize(init_points=init_round, n_iter=opt_round)
    
    model_auc=[]
    for model in range(len( lgbBO.res)):
        model_auc.append(lgbBO.res[model]['target'])
    
    # return best parameters
    return lgbBO.res[pd.Series(model_auc).idxmax()]['target'],lgbBO.res[pd.Series(model_auc).idxmax()]['params']

opt_params = bayes_parameter_opt_lgb(X, y, init_round=15, opt_round=85, n_folds=10, random_seed=6,n_estimators=100)
```

```python
opt_params[1]["num_leaves"] = int(round(opt_params[1]["num_leaves"]))
opt_params[1]['max_depth'] = int(round(opt_params[1]['max_depth']))
opt_params[1]['min_data_in_leaf'] = int(round(opt_params[1]['min_data_in_leaf']))
opt_params[1]['max_bin'] = int(round(opt_params[1]['max_bin']))
opt_params[1]['objective']='binary'
opt_params[1]['metric']='auc'
opt_params[1]['is_unbalance']=True
opt_params[1]['boost_from_average']=False
opt_params=opt_params[1]
opt_params
```

### Profit Curve

In order to attach business value to this model, let us make few assumptions:
- Lead converted into sales will yield $120 in Revenue.

- Time and efforts spent on chasing leads is $30.

If you just do a little maths here, you will arrive at $30,000 in profit. Here’s how:

| Description | Customer | Value | Total |
| --- | --- | --- | --- |
| True Positive | 340 | \$ 120 | \$ 40800 |
| True Positive + False Positive | 340 + 20 | - \$ 30 | \$ 10800 |
|  |  |  | \$ 30000 |


```python
import numpy as np

def calculate_profit(y, y_pred):
    tp = np.where((y_pred==1) & (y==1), (120-30), 0)
    fp = np.where((y_pred==1) & (y==0), -30, 0)
    #fn = np.where((y_pred==0) & (y==1), -120, 0)
    return np.sum([tp,fp])
```

```python
calculate_profit(y_test, model[1].fit(X_train,y_train).predict(X_test))
```

```python
def pc(X_train,y_train,X_test,y_test,model,func):
  model.fit(X_train,y_train)

  pc = pd.DataFrame({'yes_proba' : model.predict_proba(X_test)[:,1],
                  'yes_truth' : y_test}).sort_values(by='yes_proba',ascending=False)
  pc['yes'] = pc['yes_proba'] > 0.5
  pc['yes'] = pc['yes'].astype('int')

  import math

  profit = []
  percentage = []
  for i in range(pc.shape[0]):
    profit.append(func(pc['yes_truth'][0:i+1], pc['yes'][0:i+1]))
    percentage.append(math.floor(100*i/pc.shape[0]))

  pc['profit'] = profit
  pc['percentage'] = percentage
  return pc


pc_model = {}
for i in range(len(model)):
  pc_model[model_name[i]] = pc(X_train,y_train,X_test,y_test,model[i],calculate_profit).groupby('percentage')['profit'].max().values
```
